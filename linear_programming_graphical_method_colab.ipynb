```
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Create a range of x values
x = np.linspace(0, 20, 400)

# Calculate y values based on each constraint
y1 = 18 - 2*x
y2 = (42 - 2*x) / 3
y3 = 24 - 3*x

plt.figure(figsize=(10,10))

# Plot each constraint
plt.plot(x, y1, label=r'$2x + y \leq 18$')
plt.fill_between(x, 0, y1, where=(y1>0), color='gray', alpha=0.5)

plt.plot(x, y2, label=r'$2x + 3y \leq 42$')
plt.fill_betweenx(x, 0, (42-3*x)/2, where=((42-3*x)/2>0), color='gray', alpha=0.5)

plt.plot(x, y3, label=r'$3x + y \leq 24$')
plt.fill_between(x, 0, y3, where=(y3>0), color='gray', alpha=0.5)

# Highlight the feasible region and set the labels and title
plt.xlim((0, 15))
plt.ylim((0, 15))
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.axvline(0, color='black',linewidth=0.5)
plt.axhline(0, color='black',linewidth=0.5)
plt.grid(True, which='both')
plt.title('Feasible Region')
plt.legend()

plt.show()

# Additional code for finding the solution
def line_intersection(line1, line2):
    xdiff = (line1[0][0] - line1[1][0], line2[0][0] - line2[1][0])
    ydiff = (line1[0][1] - line1[1][1], line2[0][1] - line2[1][1])

    def det(a, b):
        return a[0] * b[1] - a[1] * b[0]

    div = det(xdiff, ydiff)
    if div == 0:
        return None

    d = (det(*line1), det(*line2))
    x = det(d, xdiff) / div
    y = det(d, ydiff) / div
    return x, y

line1 = [(0, 18), (9, 0)]
line2 = [(0, 14), (21, 0)]
line3 = [(0, 24), (8, 0)]

intersection_points = [line_intersection(line1, line2), line_intersection(line1, line3), line_intersection(line2, line3)]
z_values = [3*x + 2*y for x, y in intersection_points]
max_z = max(z_values)
max_point = intersection_points[z_values.index(max_z)]

max_z, max_point
```
